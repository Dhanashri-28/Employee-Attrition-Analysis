```python
#Data manipulation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data HR Analytics
df= pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
#to display first 5 rows
df.head()
```

```python
#Check dataset shape( Rows and columns)
df.shape
```

```python
#Understanding the data
#Data Structure, Column type, and not null values
df.info()
```

```python
#For statistical summary
df.describe()
```

```python
#Check if there is any null values
df.isnull().sum()
```

```python
#count employee who left vs stayed
df['Attrition'].value_counts()
```

```python
#Plot attrition count
sns.countplot(x= 'Attrition', data=df)
plt.title( 'Employee Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
```

```python
## Most employees stayed, but a significant number left the organisation ##
```

```python
#Univariate Analysis
#Age Distribution
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Monthly Income Distribution
sns.histplot(df['MonthlyIncome'], kde=True)
plt.title('Monthly Income distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting Categorical variables
#Job role distribution
sns.countplot(y='JobRole', data=df)
plt.title('Employee count by Job Role')
plt.xlabel('Number of Employees')
plt.ylabel('Job Role')
plt.show()
```

```python
#Department Distribution
sns.countplot(x= 'Department', data=df)
plt.title('Employee count by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#BIVARIATE ANALYSIS
#Attrition BY Department
sns.countplot(x='Department',hue='Attrition', data=df)
plt.title('Attrition BY Department')
plt.xlabel('Department')
plt.ylabel('Attrition')
plt.show()
```

```python
#Sales Department Shoes higher attrition as compared to others
```

```python
#Attrition By Job Role
sns.countplot(y='JobRole',hue='Attrition', data=df)
plt.title('Attrition by Job Role')
plt.xlabel('Number Of Employees')
plt.ylabel('Attrition')
plt.show()
```

```python
#Attrition VS Monthly Income
sns.boxplot(x='Attrition',y = 'MonthlyIncome', data=df)
plt.title('Attrition By Monthly Income')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

```python
#Attrition VS Age
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Attrition by Age')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()
#YOUNGER EMPLOYEES SHOWS HIGHER ATTRITON
```

```python
#Work Experience Analysis
#Attrition VS Years at Company
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years At company VS Attriton')
plt.xlabel('Attrition')
plt.ylabel('Years at company')
plt.show()
#EMPLLOYEES WITH FEWER YEARS AT THE COMPANY TEND TO LEAVE MORE
```

```python
#Attrition VS Job Satisfaction
sns.boxplot(x='JobSatisfaction', y='Attrition', data=df)
plt.title('Job Satisfaction VS Attriton')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#Multivariate Analysis
#select important numerical columns
corr_cols= [ 'Age', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany',
            'YearsInCurrentRole', 'YearsSinceLastPromotion']

plt.figure(figsize=(12,8))
sns.heatmap(df[corr_cols].corr(),annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap Of Numerical Features')
plt.show()
```

Conclusion:

Attrition Reults:  Younger employees, low income, short tenure, low job satisfaction and sales department have higher attrition

Employee Retention Strategy:    Offer competitive salaries and benefits, provide career growth and training, improve job satisfaction and engagement, support high-risk departments, implement mentorship programs, and monitor attrition trends to reduce employee turnover.
